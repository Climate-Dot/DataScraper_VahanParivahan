```python
import os
from pymysql import connect
import pandas as pd
import numpy as np
from datetime import datetime
import sqlalchemy
from sqlalchemy import create_engine
```

```python
data_file_folder = r'D:\Users\shubh\Documents\ClimateDot\EV\Parivahan\Data\nextmonth'
```

```python
master_df = pd.DataFrame()
df = pd.DataFrame()

for file in os.listdir(data_file_folder):
    
    if file.endswith('.xlsx'):
        print("loading file {0}. . .".format(file))
        df = pd.read_excel(os.path.join(data_file_folder, file))
        
        year = [*df][0][-5:-1] #extract year using string manipulation from the first cell of the excel
        month = [*df][0][-9:-6] #extract year using string manipulation from the first cell of the excel
        state_name = df.columns[0].split(' of ')[-1].split(' (')[0].strip()
        df['State'] = state_name

        df.iloc[2,0] = "Sr" #rename first column as 'Sr'
        df.iloc[2,1] = "Vehicle Class" #rename second column as 'Vehicle Class'
        df.iloc[2,-2] = "Total" #rename last column as 'Total'
        df.iloc[2,-1] = "State"
        new_header = df.iloc[2] #grab the first row for the header
        df = df[3:] #take the data less the header row
        df.columns = new_header #set the header row as the df header

        df['Year'] = year #add column year that specifies year
        df['Month_name'] = month #add column month that specifies month
        df.set_index("Sr",inplace = True) #drop the index column of dataframe
        
        master_df = pd.concat([master_df,df])
```

```python
master_df['Vehicle Type'] = master_df['Vehicle Class'].map(
    {v: k for k, vehicles in vehicle_type_dict.items() for v in vehicles}, na_action='ignore')
master_df
```

```python
vehicle_category_dict = {'2-Wheelers': ['2W_Personal', '2W_Shared'], 
                         '3-Wheelers': ['3W_Personal', '3W_Shared','3W_Shared_LowSpeed','3W_Goods','3W_Goods_LowSpeed'],
                        '4-Wheelers': ['4W_Personal', '4W_Shared'],
                        'Bus': ['Bus','Institution Bus'],
                        'Others': ['Others']}
```

```python
master_df['Vehicle Category'] = master_df['Vehicle Type'].map(
    {v: k for k, categories in vehicle_category_dict.items() for v in categories}, na_action='ignore')
master_df
```

```python
master_df['Month_name'] = [str(i).lower() for i in master_df['Month_name']]
master_df['Day'] = 1
master_df
```

```python
cols = ['Day', 'Month_name', 'Year']
master_df['Date'] = master_df[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
master_df
```
